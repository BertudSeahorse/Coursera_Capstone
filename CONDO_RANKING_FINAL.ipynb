# METHODS

We are gonna rank each condo by its proximity to the amenities that angel required.
here are the methods that we will use.

### DISTANCE CALCULATOR

```python
def distance_measurer(lon1, lat1, lon2, lat2):
    
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    return 6371 * (
        acos(sin(lat1) * sin(lat2) + cos(lat1) * cos(lat2) * cos(lon1 - lon2))
    )
```

### COLUMN'S DATAFRAME FINDER

```python
def column_df_finder(column):
    if column == 1 :
        df_ameni = df_gym
    else:
        if column == 2:
            df_ameni = df_grocery
        else:
            if column == 3:
                df_ameni = df_pharmacy
            else:
                if column == 4 :
                    df_ameni = df_train                
                else:
                    df_ameni = ""
                            
    return df_ameni
```

### SCORER

```python
def scorer(df_condo, df_ameni, index, row_ameni_in):
        
    row_ameni = row_ameni_in
        
    lat1 = df_condo.iloc[index , 3]
    lon1 = df_condo.iloc[index , 4]    
    
    i = len(df_ameni.index)
    
    while i > row_ameni:
                                                     
                lat2 = df_ameni.loc[row_ameni, 'lat']
                lon2 = df_ameni.loc[row_ameni, 'lng']
            
            
                distance = distance_measurer(lon1, lat1, lon2, lat2)
                
                if distance <= 0.5:
                    row_ameni = 20
                else: 
                    row_ameni = row_ameni + 1
        
    if row_ameni == 20:
        
        score = 1
        
    else: 
        
        score = 0
    
    return score 
    
    
```

### SCORE FACTOR FINDER

```python
def score_factor(column):
    if column == 1 :
        score_factor = 0.75 #gym
    else:
        if column == 2:
            score_factor = 0.75 #grocery
        else:
            if column == 3:
                score_factor = 0.5 # pharmacy
            else:
                if column == 4 :
                    score_factor = 1 # train       
                else:
                    score_factor = 1
                            
    return score_factor
```

### TOTALER

```python
def totaler(df, index, column): # index must return to 0 first and column must be 7
    while index < len(df.index):
        
        df.iloc[index,column] = (df.iloc[index, 1]) + (df.iloc[index,2]) + (df.iloc[index,3])  + (df.iloc[index,4]) 
        index = index + 1
    return df
```

# THE RANKING DATAFRAME

#### FIRST WE CREATE A BLANK DATAFRAME

```python

df = pd.DataFrame(df_condo, columns = ['name'])
df['gym'] = 0.0
df['grocery'] = 0.0
df['pharmacy'] = 0.0
df['train'] = 0.0

df['total score'] = 0.0
df
```

###  WE RANK THEM BY INPUTING THE SCORES AND SORTING THEM FROM HIGHEST TO LOWEST

```python
index = 0
column = 4
row_ameni_in = 0




while column < 5:
    
        
        df_ameni = column_df_finder(column)
    
        while index < len(df.index):
                
                score = scorer(df_condo, df_ameni, index, row_ameni_in) * score_factor(column)
            
                df.iloc[index,column] = score
        
                index = index + 1  
          
        column = column + 1
        
index = 0 
column = 5
df = totaler(df, index, column)
df

df = df.sort_values(['total score'], ascending=[False])
df
```
